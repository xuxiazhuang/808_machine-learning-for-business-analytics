```python
import keras
from keras import layers
```

```python
import numpy as np
dataset = np.loadtxt("pima-indians-diabetes.csv",delimiter = ',')
```

```python
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
y.shape
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(200,input_dim = 8,activation = "relu")) ### 200 is nurons, 8 is feature
model.add(keras.layers.Dense(150,activation = "relu"))
model.add(keras.layers.Dense(150,activation = "relu"))
model.add(keras.layers.Dense(90,activation = "relu"))
model.add(keras.layers.Dense(1,activation = "sigmoid"))### softmax for the multiple class, sigmoid is for the binary class
```

```python
### set the model to train
model.compile(loss = "binary_crossentropy",optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.fit(x,y,epochs = 150, batch_size = 10)### we can also add validation_data
```

```python
# model.predict_classes(x)
```

```python
model.fit(x,y,epochs = 10, batch_size = 20,validation_split = 0.2)### we can also add validation_data
### change batch_size to reduce overfitting , it's also a way
### here batch_size is the number of sample, the higher of the batch_size the faster of the speed.
### The more batches , learning more. learning more indicate learning more slowly.
```

```python
model.fit(x,y,epochs = 10, batch_size = 100,validation_split = 0.2)
```

```python
model.fit(x,y,epochs = 10, batch_size = 5,validation_split = 0.2)
```

```python
### Returns the loss value & metrics values for the model.
model.evaluate(x,y)### this is the measurment for the final model,the first number is loss, the second one is accracy
```
